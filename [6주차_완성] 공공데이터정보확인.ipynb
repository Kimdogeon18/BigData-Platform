```python
# 상도동 실거래가 가져오기

import pandas as pd

aptData = pd.read_excel('상도동 아파트 실거래가.xlsx', header=[16])
```

```python
aptData.head()
```

```python
# 상도동 실거래가 데이터 확인하기
aptData.info()
```

### 행 데이터 삭제
변수명.drop(index=행번호, axis=0, inplace=True) : index가 ‘행번호’인 행 삭제

변수명.drop(index=[0,2,4], axis=0): index가 0,2,4인 행(3줄)만 삭제

변수명.drop(변수명.index[0:17], axis=0): 0~16 index행 삭제

```python
data = aptData.drop(index = [0], axis = 0)
data.head(20)
```

```python
aptData.drop(aptData.index[0:17], axis = 0, inplace = True)
aptData.head()
```

### index 번호 재설정
변수명.reset_index(drop=True, inplace=True)

drop=True 옵션은 기존 인덱스는 버리고 새로 인덱스 설정

```python
aptData.reset_index()
```

```python
aptData.reset_index(drop = True, inplace=True)
aptData.head()
```

### 열 데이터 삭제
#### 변수명.drop(columns=[‘열이름’], axis=1, inplace=True) : ‘열이름’열 삭제
여러열 삭제 : 변수명.drop(columns=[‘열이름1’,’열이름2’], axis=1) : ‘열이름1’,’열이름2’열 삭제

변수명.drop(변수명.columns[0:5], axis=1): 0~4 index 열 삭제

```python
data = aptData.drop(columns = ['번지','본번','부번'],axis = 1)
data.head()
```

```python
data = aptData.drop(aptData.columns[1:4],axis = 1)
data.head()
```

### 열 데이터 삭제
#### 변수명.drop(columns=[‘열이름’], axis=1, inplace=True) : ‘열이름’열 삭제
#### inplace= True 옵션을 추가하면 원본을 변경함

```python
aptData.drop(columns=['해제사유발생일','중개사소재지'],axis = 1, inplace=True)
aptData
```

### 열이름 변경하기
#### 변수명.rename(columns = {‘열이름’:‘새로운 열이름’}, inplace= True)

```python
aptData.rename(columns={'전용면적(㎡)':'전용면적','거래금액(만원)':'거래금액'},inplace=True)
aptData.head()
```

### 데이터 열 연산 및 새로운 열 생성하기
#### 변수명[‘열이름’] = 변수명[‘열이름’] + 변수명[‘열이름‘] 

```python
aptData['평수'] = aptData['전용면적']/3.3
aptData.head()
```
