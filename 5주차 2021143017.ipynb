```python
data.rename(columns={'출생아수(명)':'출생아수','혼인건수(건)':'혼인건수'}, inplace=True)
data
```

```python
numData=data[10:]
numData
```

```python
numData.reset_index(drop=True, inplace=True)
numData.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(numData['혼인건수'])
plt.plot(numData['출생아수'])
plt.show()
```

```python
plt.plot(numData['혼인건수'], marker='o')
plt.plot(numData['출생아수'], marker='^')

plt.show()
```

```python
plt.plot(numData['혼인건수'], marker='o', label='1')
plt.plot(numData['출생아수'], marker='^', label='2')


plt.legend()
plt.show()
```

```python
plt.plot(numData['혼인건수'], marker='o', label='1')
plt.plot(numData['출생아수'], marker='^', label='2')

plt.ylim(150000, 550000)

plt.legend()
plt.show()
```

```python
plt.plot(numData['혼인건수'], marker='o', label='1')
plt.plot(numData['출생아수'], marker='^', label='2')

plt.ylim(150000, 550000)
plt.xlim(8,23)

plt.legend()
plt.show()
```

```python
plt.plot(numData['혼인건수'], marker='o', label='1')
plt.plot(numData['출생아수'], marker='^', label='2')

plt.ylim(150000, 550000)
plt.xlim(8,23)

plt.legend()
plt.title('출생아수,혼인건수')
plt.show()
```

```python
plt.plot(numData['혼인건수'], marker='o', label='1')
plt.plot(numData['출생아수'], marker='^', label='2')

plt.ylim(150000, 550000)
plt.xlim(8,23)

plt.legend()
plt.title('출생아수,혼인건수')

plt.grid(True)
plt.show()
```

```python
plt.plot(numData['혼인건수'], marker='o', label='1')
plt.plot(numData['출생아수'], marker='^', label='2')

plt.ylim(150000, 550000)
plt.xlim(8,23)
plt.legend()
plt.title('출생아수,혼인건수')
plt.grid(True)

plt.xlabel('년도')
plt.ylabel('인구')

plt.show()
```
