```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
fmri = sns.load_dataset('fmri')
titanic = sns.load_dataset('titanic')
flights = sns.load_dataset('flights')
```

```python
iris.head()
```

```python
sns.relplot(data=iris, x='sepal_length', y='petal_length');
```

```python
sns.set_style('whitegrid')
sns.relplot(data=iris, x='sepal_length', y='petal_length', hue='species')
```

```python
sns.set_style('whitegrid')
sns.relplot(data=iris, x='sepal_length', y='petal_length', kind='line')
```

```python
sns.set_style('whitegrid')
sns.relplot(data=iris, x='sepal_length', y='petal_length', hue='species', kind='line')
```
