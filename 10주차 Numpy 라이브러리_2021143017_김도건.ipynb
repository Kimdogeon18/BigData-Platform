```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 3.14*2, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label='sine')
plt.plot(x, y2, label='cosine')

plt.title('sine & cosine graph')
plt.xlabel('x value')
plt.ylabel('y value')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
scores = []
for i  in range(5) :
    val = int(input('%d  ->' % (i+1) ))
    scores.append( val )

print('\n ')
print(scores)

```

```python
sumR = np.sum(scores)
aveR = np.mean(scores)
varR = np.var(scores)
stdR = np.std(scores)

print(': %10d' %sumR)
print(': %10.2f' %aveR)
print(': %10.2f' %varR)
print(': %10.2f' %stdR)
```

```python
import numpy as np
scores = []
for i  in range(5) :
    val = int(input('%d  ->' % (i+1) ))
    scores.append( val )

print('\n ')
print(scores, '\n')
```

```python
sumR = sum(scores)
aveR = sumR / len(scores)

print(': %10d' % sumR)
print(': %10.2f' % aveR)
```

```python
r = 0 
for i in range(5) :
    r += (scores[i] - aveR)**2
    varR = r / len(scores)
    stdR = varR**(1/2)
    
print(': %10d' %sumR)
print(': %10.2f' %aveR)
print(': %10.2f' %varR)
print(': %10.2f' %stdR)
```
